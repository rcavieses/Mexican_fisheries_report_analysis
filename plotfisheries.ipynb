```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import read_csv
```

```python
data = read_csv('conapesca_2000-2015.csv', encoding='ISO-8859-1')
data.columns
```

```python
labels_state, unique_state = pd.factorize(data['Estado'])
data['labels_state'] = labels_state
```

```python
table_label_corres = pd.DataFrame({
    'State': unique_state,
    'Label': range(len(unique_state))  
})

# Mostrar la tabla de correspondencia
print(table_label_corres)
```

*  Baja California          1
*  Baja California Sur      12
*  Sonora                   0
*  Sinaloa                  4

```python
study_zone = [1, 12, 0, 4]
aggregated_data_filtered = data[data['labels_state'].isin(study_zone)]
```

```python
aggregated_data_filtered['Ano']=aggregated_data_filtered['Ano'].astype(int)
meses_a_numero = {
    'ENERO': 1, 'FEBRERO': 2, 'MARZO': 3, 'ABRIL': 4, 'MAYO': 5, 'JUNIO': 6,
    'JULIO': 7, 'AGOSTO': 8, 'SEPTIEMBRE': 9, 'OCTUBRE': 10, 'NOVIEMBRE': 11, 'DICIEMBRE': 12
}
aggregated_data_filtered.loc[:, 'Mes_Num'] = aggregated_data_filtered['Mes'].map(meses_a_numero)
aggregated_data_filtered.loc[:, 'Fecha'] = pd.to_datetime(aggregated_data_filtered['Ano'].astype(str) + '-' + aggregated_data_filtered['Mes_Num'].astype(str) + '-01')
```

```python
def assign_zone_to_offices(input_aggregated_data, zones_aggregated_data):
    """
    Assigns the corresponding zone to each office in the input DataFrame based on the zones DataFrame.

    :param input_aggregated_data: DataFrame with a column 'Office' containing office names.
    :param zones_aggregated_data: DataFrame with two columns 'Office' and 'Zone', mapping offices to their zones.
    :return: DataFrame with an additional 'Zone' column.
    """
    zone_mapping = dict(zones_aggregated_data.values)
    input_aggregated_data['Zone'] = input_aggregated_data['Oficina'].map(zone_mapping)

    return input_aggregated_data
```

```python
ofice_list = pd.read_csv('list_zones.csv')
```

```python
aggregated_data_filtered = assign_zone_to_offices(aggregated_data_filtered, ofice_list)
```

```python
# Set the year and month as a new column for grouping
aggregated_data_filtered['YearMonth'] = aggregated_data_filtered['Fecha'].dt.to_period('M')

# Grouping by 'NombrePrincipal', 'Zone', 'YearMonth' and aggregating the sum for the specified columns
# Adding a frequency column that counts the number of occurrences of each 'NombrePrincipal' per 'Zone' and 'YearMonth'
aggregated_data= aggregated_data_filtered.groupby(['NombrePrincipal', 'Zone', 'YearMonth'])\
                                     .agg({'PesoDesembarcado': 'sum', 
                                           'PesoVivo': 'sum', 
                                           'Precio': 'mean', 
                                           'Valor': 'sum',
                                           'NombrePrincipal': 'size'})\
                                     .rename(columns={'NombrePrincipal': 'Frecuencia'}).reset_index()
```

```python
aggregated_data.to_csv('aggregated_data.csv')
```

```python
# Display the first few rows of the aggregated DataFrame
def queryfisherie(PrincipalName,WichZone): 
    Consult = aggregated_data[aggregated_data['NombrePrincipal']==PrincipalName]
    Consult = Consult[Consult['Zone']==WichZone]
    return Consult

def plotfisherie(data,Name,Zone):

    # Sorting the data by YearMonth for proper time series plotting
    data_corrected = data.sort_values('YearMonth')

    # Convert 'YearMonth' to datetime for plotting
    data_corrected['YearMonth'] = data_corrected['YearMonth'].dt.to_timestamp()

    # Re-plotting the time series graph for 'PesoVivo'
    plt.figure(figsize=(30, 6))
    plt.plot(data_corrected['YearMonth'], data_corrected['PesoVivo'],'--b', marker='o')
    plt.title('Time Series for PesoVivo for '+ Name + 'en Zona '+ Zone)
    plt.xlabel('Date (Year y Month)')
    plt.ylabel('Kg')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()

    # Show the plot
    plt.show()
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd
import plotly.graph_objs as go

def plotfisherie(data, Name, Zone):
    # Ordenar los datos por YearMonth para un correcto trazado de la serie de tiempo
    data_corrected = data.sort_values('YearMonth')

    # Convertir 'YearMonth' a datetime para el trazado
    data_corrected['YearMonth'] = data_corrected['YearMonth'].dt.to_timestamp()

    # Crear una gráfica de serie de tiempo usando Plotly
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=data_corrected['YearMonth'], y=data_corrected['PesoVivo'],
                             mode='lines+markers', name='PesoVivo',
                             line=dict(color='blue', dash='dash')))
    fig.update_layout(title='Time Series for PesoVivo for ' + Name + ' in Zona ' + str(Zone),
                      xaxis_title='Date (Year and Month)',
                      yaxis_title='Kg',
                      xaxis=dict(tickformat="%Y/%m"))

    return fig

# Inicializar la aplicación Dash
app = dash.Dash(__name__)

# Esqueleto de la aplicación
app.layout = html.Div([
    dcc.Dropdown(
        id='nombre_principal_dropdown',
        options=[{'label': i, 'value': i} for i in aggregated_data['NombrePrincipal'].unique()],
        value='ABULON'  # Valor por defecto
    ),
    dcc.Dropdown(
        id='zona_dropdown',
        options=[{'label': i, 'value': i} for i in aggregated_data['Zone'].unique()],
        value=1  # Valor por defecto
    ),
    dcc.Graph(id='time_series_graph'),
    html.Button('Descargar Datos', id='btn-download'),
    dcc.Download(id='download-data')
])

# Callback para actualizar la gráfica
@app.callback(
    Output('time_series_graph', 'figure'),
    [Input('nombre_principal_dropdown', 'value'),
     Input('zona_dropdown', 'value')]
)
def update_graph(selected_nombre_principal, selected_zona):
    # Usar la función queryfisherie para filtrar los datos
    filtered_data = queryfisherie(selected_nombre_principal, selected_zona)
    # Usar la función plotfisherie para crear la gráfica
    return plotfisherie(filtered_data, selected_nombre_principal, str(selected_zona))

# Callback para descargar datos
@app.callback(
    Output('download-data', 'data'),
    [Input('btn-download', 'n_clicks'),
     Input('nombre_principal_dropdown', 'value'),
     Input('zona_dropdown', 'value')],
    prevent_initial_call=True
)
def download_data(n_clicks, selected_nombre_principal, selected_zona):
    
    filtered_aggregated_data = aggregated_data[(aggregated_data['NombrePrincipal'] == selected_nombre_principal) & (aggregated_data['Zone'] == selected_zona)]
    return dcc.send_data_frame(filtered_aggregated_data.to_csv, filename="time_series_data.csv")

# Correr el servidor
if __name__ == '__main__':
    app.run_server(debug=True)
```
